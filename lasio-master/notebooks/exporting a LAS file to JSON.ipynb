```python
import json
import os
from pprint import pprint

import lasio
```

```python
l = lasio.read(os.path.join("..", "tests", "examples", "6038187_v1.2_short.las"))
```

To export to JSON you need to use a special encoder:

```python
t = json.dumps(l, cls=lasio.JSONEncoder,  # <--- this is necessary
               sort_keys=True, indent=4, separators=(',', ': ') # <-- these are just for making it pretty below.
               )
```

Note that this string representation only contains the data - the other information will be included in the future ([issue #107](https://github.com/kinverarity1/lasio/issues/107)).
